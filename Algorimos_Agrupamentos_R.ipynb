# Agrupamento 

```r
install.packages("factoextra")
library(factoextra)
```

Gerando o kluster

```r
cluster <- kmeans(iris[, 1:4], centers = 3)
```

```r
table(iris$Species, cluster$cluster)
```

```r
# Grafico
plot(iris$Sepal.Length, iris$Sepal.Width, col = cluster$cluster, pch = 20, main = "iris")
plot(iris[, 1:4], col = cluster$cluster, main = "Iris")
```

Grafico Especializado


```r
g2 = fviz_cluster(cluster, data = iris[, 1:4], ellipse.type = "covex", ggtheme = theme_bw(), main = "cluster")
plot(g2)
```

Adicionamos o cluster aos dados originais

#Cluster Hierárquico

```r
library(factoextra)
```

Criando o cluster

```r
hc <- hcut(iris[, -5], k = 3, hc_method = "complete", stand = T)
```

```r
hc
```

Dendograma

```r
plot(hc)
```

Dendograma com o pacote factoextra

```r
g3 <- fviz_dend(hc, show_labels = FALSE, rect = TRUE, ggtheme = theme_bw(), main = "Cluster Hierárquico")
plot(g3)
```

#Cluster

```r
g4 <- fviz_cluster(hc, ellipse.type = "norm", ggtheme = theme_bw(), main = "Cluster Hierárquico")
plot(g4)
```

```r
# Comparação
plot(g3)
plot(g4)
```
